```python
import pandas as pd
import matplotlib.pyplot as plt

columns=['sepal_length','sepal_width','petal_lenth','petal_width','species']

df=pd.read_csv("iris.data",names=columns)
```

```python
print(df.head())
```

```python
#features & types
print(df.dtypes)
#all numeric ie float while species/variety is categorical(object)
```

```python
#histogram to display feature distribution

df.hist(figsize=(10,8),bins=20)
plt.show()
```

```python
df.describe()
```

```python
#box plot

import seaborn as sns
import matplotlib.pyplot as plt

for col in df.columns[:-1]:
    sns.boxplot(y=df[col])
    plt.title(col)
    plt.show()
```

```python
#Detailed Inference of Boxplots:

    #The "Box": Represents the Interquartile Range (IQR), containing the middle 50% of the data.
    #The Median: The line inside the box. If it's not centered, the data is skewed.
    #The Whiskers: These extend to show the rest of the distribution. Anything outside these lines is usually flagged as an outlier.
    #Iris Observation: In this dataset, Sepal Width is the only feature that usually shows outliers (a few points sitting way above or below the whiskers), meaning some flowers have unusually wide or narrow sepals compared to the rest of the population.

#Comparing Distributions
    #By comparing histograms and boxplots, we see that Petal Length and Width have much higher variance (spread) than Sepal dimensions. 
    #The Petals are "stretched out" across the graph because they differ wildly between species.
    #This tells a data scientist that these two features are the most important for training a machine learning model.
```
